```python
%matplotlib inline
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gloabal Variables
_FIGSIZE = (11,11) #<<< size demensions of graph (x,y)
```

# Problem I-1

>Using arrows of proper magnitude and direction, sketch each of the followng vector functions:

* (a)  $\textbf{i}y + \textbf{j}x$
* (b)  $(\textbf{i} + \textbf{j})/\sqrt{2}$
* (c)  $\textbf{i}x - \textbf{j}y$
* (d)  $\textbf{i}y$
* (e)  $\textbf{j}x$
* (f)  $(\textbf{i}y + \textbf{j}x)/\sqrt{x^2 + y^2}$
* (g)  $\textbf{i}y + \textbf{j}xy$
* (h)  $\textbf{i} + \textbf{j}y$

```python
y = x = arange(-10,10,1)   # 1D list of coords (for vector fields, don't get too dense)
(xx,yy) = meshgrid(x,y)    # 2D arrays of coords
```


### (a)  $\textbf{i}y + \textbf{j}x$

```python
(vx,vy)= (yy,xx)

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=50, angles='xy')
plt.show()
```

### (b)  $(\textbf{i} + \textbf{j})/\sqrt{2}$

```python
(vx,vy)= (sqrt(2),sqrt(2))

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=50, angles='xy')
plt.show()
```

### (c) $\textbf{i}x - \textbf{j}y$

```python
(vx,vy)= (xx,-yy)

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=50, angles='xy')
plt.show()
```

### (d)  $\textbf{i}y$

```python
(vx,vy)= (yy,yy*0)

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=50, angles='xy')
plt.show()
```

### (e)  $\textbf{j}x$

```python
(vx,vy)= (xx*0,xx)

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=50, angles='xy')
plt.show()
```

### (f)  $(\textbf{i}y + \textbf{j}x)/\sqrt{x^2 + y^2}$


```python
yalt = xalt = arange(-10,10,.9)   # 1D list of coords (for vector fields, don't get too dense)
(xxa,yya) = meshgrid(xalt,yalt)    # 2D arrays of coords
vx,vy= (yya/sqrt(yya**2 + xxa**2),xxa/sqrt(yya**2 + xxa**2))

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xxa,yya,vx,vy, 1, scale=18, angles='xy')
plt.show()
```

### (g)  $\textbf{i}y + \textbf{j}xy$

```python
vx,vy=(yy,yy*xx)

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=50, angles='xy')
plt.show()
```

### (h)  $\textbf{i} + \textbf{j}y$

```python
vx,vy=(np.ones((20,20)),yy)

fig = plt.figure(figsize=_FIGSIZE)
q = plt.quiver(xx,yy,vx,vy, 1, scale=100, angles='xy')
plt.show()
```
